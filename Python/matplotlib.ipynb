```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DataSet.csv")
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
df # to drop the column named "Unnamed: 0" from the dataframe and update the dataframe in place.
```

```python
plt.hist(df['Salary'], bins = 10)
plt.xlabel('Salary')
plt.ylabel('Frequency') 
plt.title('Distribution of Salary')
plt.grid()
plt.show()

# In this code, we are creating a histogram of the 'Salary' column from the dataframe 'df' with 10 bins. We are also labeling the x-axis as 'Salary', the y-axis as 'Frequency', and giving the plot a title 'Distribution of Salary'. Finally, we are adding a grid to the plot and displaying it using plt.show().
```

```python
size = [20, 30, 40, 10, 50]
labels = ['A', 'B', 'C', 'D', 'E']

plt.pie(size, labels = labels, autopct='%1.1f%%')
plt.title('Distribution of Categories')
plt.show()

# In this code, we are creating a pie chart using the plt.pie() function. The 'size' variable contains the values for each category, and the 'labels' variable contains the corresponding labels for each category. The 'autopct' parameter is set to '%1.1f%%' to display the percentage of each category in the pie chart. Finally, we are giving the plot a title 'Distribution of Categories' and displaying it using plt.show().
```

```python
plt.boxplot(df['Salary'])
plt.xlabel('Salary')
plt.title('Box Plot of Salary')
plt.grid()      
plt.show()  

# In this code, we are creating a box plot of the 'Salary' column from the dataframe 'df' using the plt.boxplot() function. We are labeling the x-axis as 'Salary' and giving the plot a title 'Box Plot of Salary'. Finally, we are adding a grid to the plot and displaying it using plt.show().
```

```python
plt.scatter(df['Years of Experience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Salary vs. Years of Experience')
plt.grid()  
plt.show()

# In this code, we are creating a scatter plot to visualize the relationship between 'Years of Experience' and 'Salary' from the dataframe 'df'. We are labeling the x-axis as 'Years of Experience', the y-axis as 'Salary', and giving the plot a title 'Scatter Plot of Salary vs. Years of Experience'. Finally, we are adding a grid to the plot and displaying it using plt.show().
```

```python
plt.subplot (1, 2, 1)
plt.boxplot(df['Salary'])
plt.xlabel('Salary')
plt.title('Box Plot of Salary')
plt.grid()  

plt.subplot (1, 2, 2)
plt.scatter(df['Years of Experience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Salary vs. Years of Experience')
plt.grid()

# In this code, we are creating a figure with two subplots using the plt.subplot() function. The first subplot is a box plot of the 'Salary' column from the dataframe 'df', and the second subplot is a scatter plot to visualize the relationship between 'Years of Experience' and 'Salary'. We are labeling the axes and giving each plot a title. Finally, we are adding a grid to both plots for better visualization.
```

```python
plt.bar(df['Years of Experience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Bar Plot of Salary vs. Years of Experience')
plt.grid()
plt.show()

# In this code, we are creating a bar plot to visualize the relationship between 'Years of Experience' and 'Salary' from the dataframe 'df'. We are labeling the x-axis as 'Years of Experience', the y-axis as 'Salary', and giving the plot a title 'Bar Plot of Salary vs. Years of Experience'. Finally, we are adding a grid to the plot and displaying it using plt.show().
```
